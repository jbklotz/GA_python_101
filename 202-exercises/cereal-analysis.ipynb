## 7) Which five cereals have the highest sugar content per _cup_?

```python
cereal.sort_values("sugars", ascending =False).loc[:, ["name", "sugars", "cups"]].head()
```

```python
#make a new column
cereal['sugar_per_cup']= cereal.sugars / cereal.cups
```

```python
cereal.head()
```

```python
cereal.sort_values("sugar_per_cup", ascending =False).loc[:, ["name", "sugar_per_cup"]].round(2).head()
```

```python
#same as above, just cleaner 
(cereal
 .sort_values("sugar_per_cup", ascending =False)
 .loc[:, ["name", "sugar_per_cup"]]
 .round(2)
 .head()
)
```
